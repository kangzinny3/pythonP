```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = 'https://search.naver.com/search.naver?sm=top_hty&fbm=1&ie=utf8&query=%ED%99%98%EC%9C%A8'
page = urllib.request.urlopen(url)
bs = BeautifulSoup(page, 'html.parser')    # BeautifulSoup 객체로 만들기
```

```python
bs
```

```python
type(bs)
```

```python
result = bs.select('span > strong')
result
```

```python
result = bs.select('span.spt_con > strong')   # 클래스 값이 spt_con인 span태그
result
```

```python
result = bs.select('span.spt_con > strong')  
result[0].string          # 리스트이므로 인덱스로 접근
```
