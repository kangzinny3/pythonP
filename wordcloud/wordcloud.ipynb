```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline          # jupyter lab에서 바로 실행시킬 수 있도록
```

```python
stopwords = []
```

```python
def displayWordCloud(data = None, backgroundcolor = 'white', width=800, height=6000):
    wordcloud = WordCloud(
                        font_path = './NanumPen.ttf',
                        # stopwords = stopwords_kr,    # 제외시키고 싶은 단어 (리스트로 미리 만들어두기)
                        background_color = backgroundcolor,
                        width = width, height = height).generate(data)
    print(wordcloud.words_)
    plt.figure(figsize = (15, 10))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
%time displayWordCloud(story)
```

```python
from soynlp.noun import NewsNounExtractor
```

```python
%%time
noun_extractor = NewsNounExtractor()
nouns = noun_extractor.train_extract([story])    # nouns에 story에서 추출된 명사를 넣음
                                                 # .train_extract() 안에는 리스트가 들어가야함
```

```python
print(type(nouns))
```

```python
%time displayWordCloud(' '.join(nouns))
```

```python
# 특정 이미지 형태로 워드 클라우드 그리기
```

```python
from PIL import Image
import numpy
```

```python
img = Image.open('cloud.png')   # open 메서드로 이미지 열기
img_array=numpy.array(img)
```

```python
wordcloud = WordCloud(font_path = './NanumPen.ttf', 
                      # stopwords = stopwords_kr, 
                      background_color = 'white', 
                      mask=img_array,    # mask에 이미지를 넣어줌
                      width = 800, height = 600).generate(' '.join(nouns))
plt.figure(figsize = (15 , 10)) 
plt.imshow(wordcloud)
plt.axis("off") 
plt.show()
```

```python
# 이미지로 결과 저장
wordcloud.to_file("sample.png")
```
